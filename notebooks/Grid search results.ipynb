```python
''' Initial KNN grid search '''

import pandas as pd
import pickle

result_file_template = "search_results_{}.pickle"

image_sizes = [15, 20, 25, 30]
for image_size in image_sizes: 
    
    print("Results for input images size : ({0} X {0})".format(image_size))
    
    with open(result_file_template.format(image_size), 'rb') as f_handle:
        data_dict = pickle.load(f_handle)
        result_frame = pd.DataFrame.from_dict(data_dict)
        result_frame = result_frame[['params','split0_test_score', 'split1_test_score', 'split2_test_score', 'rank_test_score']]
        result_frame = result_frame.sort_values(by=['rank_test_score'])[ : 10]
        display(result_frame)
        
    for i in range(10):
        print("Best params : ", result_frame.iloc[i]["params"])
```
